```python
from simpleai.search import SearchProblem, astar

GOAL = (10, 10, 10)

triangulo = [0, 0, 0, 0, 0, 0]
triangulo2 = [0, 0, 0, 0, 0, 0]

numero = 0



class TrianguloMagico(SearchProblem):
        
    def actions(self, state):
        global numero
        numero = numero + 1
        global triangulo
        global triangulo2
        actions = []
        if numero < 7:
            if triangulo[0] == 0:
                triangulo2[0] = numero
                actions.append(triangulo2)
                triangulo2 = [0,0,0,0,0,0]
            if triangulo[1] == 0:
                triangulo2[1] = numero
                actions.append(triangulo2)
                triangulo2 = [0,0,0,0,0,0]
            if triangulo[2] == 0:
                triangulo2[2] = numero
                actions.append(triangulo2)
                triangulo2 = [0,0,0,0,0,0]
            if triangulo[3] == 0:
                triangulo2[3] = numero
                actions.append(triangulo2)
                triangulo2 = [0,0,0,0,0,0]
            if triangulo[4] == 0:
                triangulo2[4] = numero
                actions.append(triangulo2)
                triangulo2 = [0,0,0,0,0,0]
            if triangulo[5] == 0:
                triangulo2[5] = numero
                actions.append(triangulo2)
                triangulo2 = [0,0,0,0,0,0]
                
            print(actions)
            
        return actions
        

    def result(self, state, action):
        print(state)
        triangulo[0] += action[0] 
        triangulo[1] += action[1]
        triangulo[2] += action[2]
        triangulo[3] += action[3]
        triangulo[4] += action[4]
        triangulo[5] += action[5]
        print(triangulo)
        state = (triangulo[0] + triangulo[1] + triangulo[2],triangulo[2] + triangulo[3] + triangulo[4],triangulo[4] + triangulo[5] + triangulo[0])
        return state

    def is_goal(self, state):
        return state == GOAL

    

problem = TrianguloMagico(initial_state= (0, 0, 0))
result = astar(problem)

print(result.state)
print(result.path())
```
